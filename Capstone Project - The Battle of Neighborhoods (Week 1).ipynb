```python
toronto_data = pd.read_csv("https://www.toronto.ca/ext/open_data/catalog/data_set_files/2016_neighbourhood_profiles.csv", encoding="latin1")
print("Data has been loaded!")
```

**Now, Explore the data, we have collected:**

```python
toronto_data.head()
```

**Now, we will collect the Neighbourhood Names**

```python
neigh_list = list(toronto_data.columns.values)[5:] #use indexing to remove uncessary data
neigh_list[:25]
```

**Now, we will collect neighbourhood population**

```python
#First Creating new dataset

population_data = pd.DataFrame(index=neigh_list, columns=["Population", "Income"])
population_data.head()
```

**Now, Filling the data in population_data dataframe**

```python
for index, row in population_data.iterrows():
    population_data.at[index, "Population"] = toronto_data[index][2]
    population_data.at[index, "Income"] = toronto_data[index][2264]
    
#printing the data
population_data.head()
```
